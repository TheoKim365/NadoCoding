## 17. Matplotlib 퀴즈

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic'  # Windows
# matplotlib.rcParams['font.family'] = 'AppleGothic'  # Mac
matplotlib.rcParams['font.size'] = 15   # 글자 크기
matplotlib.rcParams['axes.unicode_minus'] = False  # 한글 폰트 사용시, 마이너스 글자 깨짐현상을 해결

data = {
    '영화' : ['명량', '극한직업', '신과함께-죄와 벌', '국제시장', '괴물', '도둑들', '7번방의 선물', '암살'],
    '개봉 연도' : [2014, 2019, 2017, 2014, 2006, 2012, 2013, 2015],
    '관객 수' : [1761, 1626, 1441, 1426, 1301, 1298, 1281, 1270], # (단위 : 만 명)
    '평점' : [8.88, 9.20, 8.73, 9.16, 8.62, 7.64, 8.83, 9.10]
}
df = pd.DataFrame(data)
df
```

```python
labels = df['영화']
values = df['평점']

plt.bar(labels, values)
```

```python
labels = df['영화']
values = df['평점']

plt.bar(labels, values)   # x, y
plt.title('국내 Top 8 영화 평점 정보')
plt.xlabel('영화')
plt.ylabel('평점')
plt.xticks(rotation= 90)  # X축이 이름 데이터 각도를 90도로 설정 
plt.show()
```

```python
df_group = df.groupby('개봉 연도').mean()
df_group
```

```python
plt.plot(df_group.index, df_group['평점'])
# plt.plot(df_group['평점'])
# plt.xticks(df['개봉 연도'])
plt.xlabel('개봉 연도')
plt.ylabel('평균 평점')
plt.legend()
plt.show()
```

```python
plt.plot(df_group.index, df_group['평점'], marker = 'o')
# plt.plot(df.group['평점'], marker='o')
plt.xticks([2005, 2010, 2015, 2020])
plt.ylim(7, 10)
plt.legend()
plt.xlabel('개봉 연도')
plt.ylabel('평균 평점')
plt.show()
```

```python
plt.pie(values, labels= labels, autopct='%.1f%%')  #, startangle=90, counterclock=False)

plt.legend(loc=(1.2, 0.1))  #'right')
plt.show()
```

```python
filt = df['평점'] >= 9.0
values = [len(df[filt]), len(df[~filt])]
labels = ['9점 이상','9점 이하']

plt.pie(values, labels=labels, autopct='%.1f%%')
plt.legend(loc=(1, 0.5))
plt.show()
```
