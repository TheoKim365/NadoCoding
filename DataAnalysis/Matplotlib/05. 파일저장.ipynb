## 05. 파일저장

```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic'  # Windows
# matplotlib.rcParams['font.family'] = 'AppleGothic'  # Mac
matplotlib.rcParams['font.size'] = 15   # 글자 크기
matplotlib.rcParams['axes.unicode_minus'] = False  # 한글 폰트를 사용시, 마이너스 글자 깨짐현상을 해결
```

```python
x = [1, 2, 3]
y = [2, 4, 8]
```

```python
plt.plot(x, y)

plt.savefig('graph.png', dpi=100)
```

```python
plt.figure(dpi=200)
plt.plot(x, y)
```

```python
plt.figure(dpi=200)   # dpi : 화면에 보여지는 값
plt.plot(x, y)
plt.savefig('graph200.png')   # 저장시에도 동일한 dpi 사용
```

```python
plt.figure(dpi=200)   # dpi : 화면에 보여지는 값
plt.plot(x, y)
plt.savefig('graph100.png', dpi=100, facecolor='white')   # VS code에서는 facecolor='white' 지정해야 파일에서 축이 출력됨
```
