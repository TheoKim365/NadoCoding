## 02. 축

```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic'  # Windows
# matplotlib.rcParams['font.family'] = 'AppleGothic'  # Mac
matplotlib.rcParams['font.size'] = 15   # 글자 크기
matplotlib.rcParams['axes.unicode_minus'] = False  # 한글 폰트를 사용시, 마이너스 글자 깨짐현상을 해결
```

```python
x = [1, 2, 3]
y = [2, 4, 6]
```

```python
plt.plot(x, y)
plt.title('꺽은선 그래프', fontdict = {'family': 'HYGungso-Bold','size':20, 'color': 'red'})  # 개별 폰트 설정
```

```python
plt.plot(x, y)
plt.xlabel('X축')
plt.ylabel('Y축')
```

```python
plt.plot(x, y)
plt.xlabel('X축', color='red', size=12, loc='right')  # left, center, right
plt.ylabel('Y축', color='#00aa00', size=12, loc='top') # top, center, bottom
```

```python
plt.plot(x, y)
plt.xticks([1, 2, 3])
plt.yticks([3, 6, 9, 12])
plt.show()
```
