## 04. 스타일

```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic'  # Windows
# matplotlib.rcParams['font.family'] = 'AppleGothic'  # Mac
matplotlib.rcParams['font.size'] = 15   # 글자 크기
matplotlib.rcParams['axes.unicode_minus'] = False  # 한글 폰트를 사용시, 마이너스 글자 깨짐현상을 해결
```

```python
x = [1, 2, 3]
y = [2, 4, 8]
```

```python
plt.plot(x, y)
```

## Markers

```python
plt.plot(x, y, linewidth=3, marker='o')   # 구글 Matplotlib markers 검색
```

```python
plt.plot(x, y, marker='o', linestyle='None')
```

```python
plt.plot(x, y, marker='v')
```

```python
plt.plot(x, y, marker='v', markersize=10)
```

```python
plt.plot(x, y, marker='X', markersize=10)
```

```python
plt.plot(x, y, marker='3', markersize=20)   # 구글에서 Matplotlib markers 검색
```

```python
plt.plot(x, y, marker='o', markersize=15, markeredgecolor='red')   # 구글에서 Matplotlib markers 검색
```

```python
plt.plot(x, y, marker='o', markersize=15, markeredgecolor='red', markerfacecolor='yellow')   
# 구글에서 Matplotlib markers 검색
```

## 선 스타일

```python
plt.plot(x, y, linestyle=':')  # 구글 Matplotlie linestyle 검색
```

```python
plt.plot(x, y, linestyle='--')
```

```python
plt.plot(x, y, linestyle='-.')
```

```python
plt.plot(x, y, linestyle='-')  # solid line
```

## 색깔

```python
plt.plot(x, y, linestyle='--', color= 'pink')  # 참고 : Matplotlib color - List of named colors
```

```python
plt.plot(x, y, color= 'r')    # base color : r, g, b
```

```python
plt.plot(x, y, color= '#ff0000')   # RGB
```

```python
plt.plot(x, y, color= 'g')
```

```python
plt.plot(x, y, color= '#00ff00')   # RGB
```

```python
plt.plot(x, y, color= 'b')
```

```python
plt.plot(x, y, color= '#0000ff')  # RGB
```

## 포맷

```python
plt.plot(x, y, 'ro--')   # color, marker, linestyle
```

```python
plt.plot(x, y, 'bv:')
```

```python
plt.plot(x, y, 'go') 
```

```python
plt.plot(x, y, color='g', marker='o', linestyle='None')
```

## 축약어

```python
plt.plot(x, y, marker='o', mfc='red', ms=10, mec='blue', ls=':')
```

```python
plt.plot(x, y, marker='o', ms=10, mfc='red', mec='blue', ls=':', c='r')   # 구글에서 Matplotlib markerfacecolor mfc로 검색
```

## 투명도

```python
plt.plot(x, y, marker='o', ms=10, mfc='red', alpha=0.3)   # alpha : 투명도 (0 ~ 1)
```

```python
plt.plot(x, y, marker='o', ms=10, mfc='red', alpha=1)   # alpha : 투명도 (0 ~ 1)
```

## 그래프 크기

```python
plt.figure(figsize=(10,5))
plt.plot(x, y)
```

```python
plt.figure(figsize=(5, 10))
plt.plot(x, y)
```

```python
plt.figure(figsize=(10,5), dpi=150)
plt.plot(x, y)
```

## 배경색

```python
plt.figure(facecolor='yellow')
plt.plot(x, y)
```

```python
plt.figure(facecolor='#83a9e6')   # 구글 color picker 검색 
plt.plot(x, y)
```
