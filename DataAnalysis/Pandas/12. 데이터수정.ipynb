## 12. 데이터 수정

```python
import pandas as pd
df = pd.read_excel('score.xlsx', index_col='지원번호')
df
```

```python
# filt = df['국어'].astype(str).str.len() != 0
# df['새로'] = df.apply(lambda x: str(x['국어'])+'b' if len(x) !=0 else 'b', axis=1)
df['새로운'] = df.apply(lambda x: str(x['국어'])+'b' if x['국어'] !=0 else 'b', axis=1)
# assign은 if 조건을 사용하지 못함 
df
```

```python
# df = df.assign(합 =lambda x: x['영어'] + x['수학'])
df['소계'] = df.apply(lambda x: (x['영어'] + x['수학']) if x['국어'] > 80 else '', axis=1)
df
```

```python
df
```

```python
df = df.assign(Percentage = lambda x: (x['Total_Marks'] /500 * 100))
```

```python
df[df.iloc[:,3] * 0 != 0].index
```
