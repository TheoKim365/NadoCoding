```python
data = {
    '이름': ['채치수','정대만','송태섭','서태웅','강백호','변덕규','황태산','윤대협'],
    '학교': ['북산고','북산고','북산고','북산고','북산고','능남고','능남고','능남고'],
    '키': [197,184,168,187,188,202,188,190],
    '국어': [90,40,80,40,15,80,55,100],
    '영어': [85,35,75,60,20,100,65,85],
    '수학': [100,50,70,70,10,95,45,90],
    '과학': [95,55,80,75,35,85,40,95],
    '사회': [85,25,75,80,10,80,35,95],
    'SW특기': ['Python','Java','Javascript','','','C','PYTHON','C#']
}

data
```

## DataFrame 객체 생성

```python
import pandas as pd
df = pd.DataFrame(data)
df
```

## DataFrame 객체 생성 (index 생성)

```python
df = pd.DataFrame(data, index=['1번','2번','3번','4번','5번','6번','7번','8번'])
df
```

## DataFrame 객체 생성 (column 지정)
데이터 중에서 원하는 column만 선택하거나, 순서 변경 가능

```python
df = pd.DataFrame(data, columns=['이름','학교','키'])  # 원하는 column만 
df
```

```python
df = pd.DataFrame(data, columns=['이름','키','학교'])  # 순서 변경 
df
```

```python
import numpy as np
import pandas as pd

df = pd.DataFrame(np.random.rand(10, 5), columns = ['0', '1', '2', '3', '4'])
df['mean'] = df.mean(axis=1)
cols =['mean','0', '1', '2', '3', '4']
df = df[cols]
df
```

```python
import numpy as np
import pandas as pd

df = pd.DataFrame(np.random.rand(10, 5))
df['mean'] = df.mean(axis=1)
cols = df.columns.to_list()   # ['0', '1', '2', '3', '4','mean']
cols = cols[-1:] + cols[:-1]  # ['mean','0', '1', '2', '3', '4']
df = df[cols]
df
```
